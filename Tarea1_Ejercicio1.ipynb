```python
import numpy as np
import random

# Crear matrices grandes aleatorias de tamaño nxn
n = random.randint(10, 100)  # Tamaño de la matriz
tamaño = 10  # Tamaño de bloque

# Generamos matrices A y B con valores aleatorios
A = np.random.randint(0, 10, (n, n)).tolist()
B = np.random.randint(0, 10, (n, n)).tolist()

print(f"Tamaño de las matrices: {n}x{n}")
print(f"\nMatriz A:\n {A}")
print(f"\nMatriz B:\n {B}")
```

Multiplicacion de matrices por bloques

```python
def multiplicar_por_bloques(A, B, tamaño):
    """
    Multiplica dos matrices A y B por bloques
    """
    n = len(A)  # Asumimos matrices cuadradas de tamaño nxn
    C = [[0] * n for _ in range(n)] # Matriz resultado inicializada en ceros, garantizas el mismo tamaño que A y B

    # Recorrer la matriz en bloques
    for i in range(0, n, tamaño):  # Filas
        for j in range(0, n, tamaño):  # Columnas
            for k in range(0, n, tamaño):  # Bloques intermedios
                # Multiplicación de bloques
                A_bloque = [row[k:k+tamaño] for row in A[i:i+tamaño]]
                B_bloque = [row[j:j+tamaño] for row in B[k:k+tamaño]]

                # Verificar si los bloques están vacíos usando len()
                if len(A_bloque) > 0 and len(B_bloque) > 0 and len(A_bloque[0]) > 0 and len(B_bloque[0]) > 0:
                    # Multiplicar bloques y sumar a C
                    for fila in range(len(A_bloque)):
                        for columna in range(len(B_bloque[0])):
                            for l in range(len(B_bloque)):
                                C[i + fila][j + columna] += A_bloque[fila][l] * B_bloque[l][columna]

    return C

C = multiplicar_por_bloques(A, B, tamaño)

# Mostrar matriz resultante
print("\nMatriz resultante A*B:")
print(C)
```

Suma de matrices por bloques.

```python

def suma_por_bloques(A, B, tamaño):
    """
    Suma dos matrices A y B por bloques de tamaño block_size.
    """
    n = len(A)  # Suponemos que A y B son matrices cuadradas de tamaño nxn
    D = np.zeros((n, n))  # Matriz resultado inicializada en ceros

    # Recorrer la matriz por bloques
    for i in range(0, n, tamaño):  # Filas en bloques
        for j in range(0, n, tamaño):  # Columnas en bloques
            # Se suma el bloque correspondiente
            D[i:i+tamaño, j:j+tamaño] = A[i:i+tamaño, j:j+tamaño] + B[i:i+tamaño, j:j+tamaño]

    return D

D = multiplicar_por_bloques(A, B, tamaño)

# Mostrar matriz resultante
print("\nMatriz resultante A+B:")
print(D)

```

Producto Punto de 2 vectores

```python
vecA = np.random.randint(0,10,100).tolist()
vecB = np.random.randint(0,10,100).tolist()
tamaño1 = 10

print(f"\nVector A:\n {vecA}")
print(f"\nVector B:\n {vecB}")
```

```python
def producto_punto_por_bloques(vecA, vecB, tamaño1):
    """
    Calcula el producto punto de dos vectores A y B usando bloques de tamaño
    """
    if len(vecA) != len(vecB):
        raise ValueError("Los vectores deben tener la misma longitud.")

    n = len(vecA)
    vecC = 0  # Inicializar el resultado en 0

    # Recorrer el vector en bloques
    for i in range(0, n, tamaño1):
        # Índice final del bloque, asegurando que no supere n
        indice = min(i + tamaño1, n)

        # Producto punto parcial del bloque actual
        ProductoPuntoBloque = np.dot(vecA[i:indice], vecB[i:indice])

        # Acumulamos el resultado parcial
        vecC += ProductoPuntoBloque

    return vecC

vecC = producto_punto_por_bloques(vecA, vecB, tamaño1)

print(f"\nVector C:")
print(vecC)
```

Matriz transpuesta conjugada de una matriz
compleja

```python
def matriz_compleja(n, m, rango=(-10,10)):
  """
  Generar matriz n*m de numeros aleatorios complejos
  """
  return [[complex(random.randint(*rango), random.randint(*rango)) for i in range(m)] for j in range(n)]

#Generar matriz n*m
n, m = 5,5
matriz = matriz_compleja(n,m)

print(f"\nMatriz compleja:\n {matriz}")
```

```python
def transponer_matriz_compleja(matriz):
    """
    Primero devolveremos la transpuesta de una matriz
    """
    return [[fila[i] for fila in matriz] for i in range(len(matriz[0]))]

matriz_transpuesta = transponer_matriz_compleja(matriz)

print("\nMatriz Compleja:")
for fila in matriz:
    print(fila)

print("\nMatriz Compleja Transpuesta:")
for fila in matriz_transpuesta:
    print(fila)
```

```python
def matriz_transpuesta_conjugada(matriz_transpuesta):
  """
  Luego de obtener la matriz transpuesta de nuestra matriz compleja
  obtendremos la conjugada
  """

  #return np.conjugate(matriz_transpuesta)
  return [[elemento.conjugate() for elemento in fila] for fila in matriz_transpuesta]

matrizT_conjugada = matriz_transpuesta_conjugada(matriz_transpuesta)

print("\nMatriz Compleja Transpuesta Conjugada:")
for fila in matrizT_conjugada:
    print(fila)


```

Potencia de una matriz de tamaño 50×50.

```python
#Crearemos una matriz 50*50
n = 50  # Tamaño de la matriz
tamaño = 10  # Tamaño de bloque
potencia = 5

# Generamos matrices A y B con valores aleatorios
Mat1 = np.random.randint(0, 10, (n, n)).tolist()

print(f"Tamaño de la matriz: {n}x{n}")
print(f"\nMatriz:\n {Mat1}")
```

```python
def potencia_matriz(Mat1, potencia, n):
    """
    Calcularemos la potencia de nuestra matriz usando la multiplicación por bloques, que ya habiamos realizado
    """
    resultado = [[1 if i == j else 0 for j in range(n)] for i in range(n)]  # Matriz identidad
    base = Mat1  # Matriz base para multiplicaciones sucesivas

    while potencia > 0:
        if potencia % 2 == 1:  # Si el exponente es impar
            resultado = multiplicar_por_bloques(resultado, base, n)
        base = multiplicar_por_bloques(base, base, n)  # Cuadrado de la matriz
        potencia //= 2  # Reducimos el exponente a la mitad

    return resultado

resultado = potencia_matriz(Mat1, potencia, n)

print("\nMatriz Potencia:")
print(resultado)
```

Producto cruz entre dos vectores
tridimensionales

```python
def Producto_Cruz(a, b):
    Producto_Cruz = [a[1]*b[2] - a[2]*b[1],
                     a[2]*b[0] - a[0]*b[2],
                     a[0]*b[1] - a[1]*b[0]]
    return Producto_Cruz

a = np.random.randint(0,10,3).tolist()
b = np.random.randint(0,10,3).tolist()

print(f"\nVector a:\n {a}")
print(f"\nVector b:\n {b}")
print(f"\nProducto Cruz:")
print(Producto_Cruz(a, b))
```

Rotación de un vector bidimensional mediante una matriz de rotación.

```python
import math

angulo = math.radians(90)
vector = np.random.randint(0,10,2).tolist()

def MatrizRot_2(theta):
  """
  Contruiremos la mtriz de rotación bidimensional
  """
  return [[math.cos(theta), -math.sin(theta)],
          [math.sin(theta), math.cos(theta)]]

MatrizRot2 = MatrizRot_2(angulo)

def Rotacion_Vector(MatrizRot2, vector):
  """
  Realizaremos la rotación del vector
  """
  return [MatrizRot2[0][0]*vector[0] + MatrizRot2[0][1]*vector[1],
          MatrizRot2[1][0]*vector[0] + MatrizRot2[1][1]*vector[1]]

Vector_Rotado = Rotacion_Vector(MatrizRot2, vector)

print(f"\nVector a rotar:\n {vector}")
print(f"\nVector rotado:\n {Vector_Rotado}")
```

Rotacion de un vector tridimensional mediante una matriz de rotacion.

```python
vec = np.random.randint(0,10,3).tolist()
eje = ['x', 'y', 'z']
angulo_ = math.radians(90)

def MatrizRot_3(eje, theta):
    """
    Crearemos la matriz de rotación tridimensional alrededor del eje x,y o z especificado
    """
    cos = math.cos(theta)
    sin = math.sin(theta)

    if eje == 'x':  # Rotación alrededor del eje X
        return [
            [1, 0, 0],
            [0, cos, -sin],
            [0, sin, cos]
        ]
    elif eje == 'y':  # Rotación alrededor del eje Y
        return [
            [cos, 0, sin],
            [0, 1, 0],
            [-sin, 0, cos]
        ]
    elif eje == 'z':  # Rotación alrededor del eje Z
        return [
            [cos, -sin, 0],
            [sin, cos, 0],
            [0, 0, 1]
        ]
    else:
        raise ValueError("El eje debe ser 'x', 'y' o 'z'.")

MatrizRot3 = MatrizRot_3(eje[2], angulo_)


def Rot_Vector(M, V):
    """
    Realizaremos la rotación del vector
    """
    return [
        M[0][0] * V[0] + M[0][1] * V[1] + M[0][2] * V[2],
        M[1][0] * V[0] + M[1][1] * V[1] + M[1][2] * V[2],
        M[2][0] * V[0] + M[2][1] * V[1] + M[2][2] * V[2]
    ]

vector_rotado = Rot_Vector(MatrizRot3, vec)

print(f"\nVector a rotar:\n {vec}")
print(f"\nVector rotado:\n {vector_rotado}")
```

Norma 1


```python
n=random.randint(2,5)
Matriz = np.random.randint(-1, 10, (n, n)).tolist()

def norma_1(matriz):
    """
    Calcularemos la norma 1 de una matriz
    """
    columnas = len(matriz[0])  # Número de columnas
    suma_columnas = [0] * columnas  # Lista para almacenar sumas de cada columna

    for fila in matriz:  # Recorrer filas
        for j in range(columnas):  # Recorrer columnas
            suma_columnas[j] += abs(fila[j])  # Acumular valores absolutos en cada columna

    return max(suma_columnas)  # Devolver la máxima suma de columna


Norma1 = norma_1(Matriz)

print(f"\nMatriz:\n {Matriz}")
print(f"\nNorma 1 de la matriz:\n {Norma1}")
```

Norma infinito

```python
def norma_infinito(matriz):
  """
  Clculemos la norma infinito para una matriz
  """
  suma_filas = [sum(abs(x) for x in fila) for fila in matriz]
  return max(suma_filas)

NormaInfinito = norma_infinito(Matriz)

print(f"\nMatriz:\n {Matriz}")
print(f"\nNorma Infinito de la matriz:\n {NormaInfinito}")
```

Norma 2

```python
def norma_2(matriz):
  """
  Calcularemos la norma 2 de una matriz
  """
  suma_cuadrados = sum(x**2 for fila in matriz for x in fila)
  return math.sqrt(suma_cuadrados)

Norma2 = norma_2(Matriz)

print(f"\nMatriz:\n {Matriz}")
print(f"\nNorma Infinito de la matriz:\n {Norma2}")
```
